```r
library(tidyverse)
library(data.table)
```

```r
# Load the cohort and the interval times -- interval times derived in 'create_action_space.ipynb'
data_path <- ''
labs_df <- read_csv(paste0(data_path, 'labs_cohort.csv'))
```

```r
head(labs_df)
```

```r
labs_df2 <- labs_df %>% spread(lab_id, valuenum)
```

```r
library("reshape2")
labs_df2 = dcast(data = labs_df, 
                           formula = subject_id + charttime ~ lab_id, 
                           value.var = "valuenum",mean)
```
